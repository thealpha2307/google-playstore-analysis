```python
# importing libraries
import pandas as pd
import plotly.express as px
from datetime import datetime
```

```python
# load data
df = pd.read_csv('playstore data.csv')
df.head()
```

```python
# clean
df = df.drop_duplicates()
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df['Size'] = df['Size'].str.replace('M','').str.replace('k','')
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')
print(df.shape)
```

```python
# filters
df_filtered = df[
    (df['Rating'] >= 4.0) &
    (df['Size'] >= 10)
]
print(df_filtered.shape)
```

```python
# top 10 categories 
top10 = df_filtered.groupby('Category').agg(
    Avg_Rating=('Rating','mean'),
    Total_Reviews=('Reviews','sum')
).reset_index().sort_values('Total_Reviews', ascending=False).head(10)

print(top10)
```

```python
#plot
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Bar(name='Total Reviews', x=top10['Category'], y=top10['Total_Reviews'], marker_color='steelblue'),
    secondary_y=False
)

fig.add_trace(
    go.Scatter(name='Avg Rating', x=top10['Category'], y=top10['Avg_Rating'], 
               mode='lines+markers', marker=dict(color='red', size=10), line=dict(width=3)),
    secondary_y=True
)

fig.update_layout(title='Top 10 App Categories - Rating vs Reviews')
fig.update_yaxes(title_text="Total Reviews", secondary_y=False)
fig.update_yaxes(title_text="Average Rating", secondary_y=True, range=[3.5, 5.0])

fig.show()
```
